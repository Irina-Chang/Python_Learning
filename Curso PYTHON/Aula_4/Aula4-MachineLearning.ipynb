# Projeto de inteligência artificial - Machine Learning

```python
import yfinance as yf
```

```python
codigo = input("Digite o código da ação: ")
dados = yf.Ticker(codigo).history("2y")
dados.head()
```

```python
dados.shape
```

## Tratamento de dados


```python
#resetar o índice
dados.reset_index()
```

```python
#guardando a tabela resetada em outra variável
treinamento = dados.reset_index()
```

```python
treinamento
```

```python
# usando treinamento.Date vc acessa só uma coluna, para acessar mais colunas, podemos usar a opção abaixo
treinamento["Date"]
```

```python
#acessando duas colunas
treinamento[["Date", "Close"]]
```

```python
#extraindo apenas a data da coluna Date, sem as horas
treinamento["Date"]=treinamento["Date"].dt.date
treinamento
```

```python
#filtrando os meus dados
colunas=["Date", "Close"]
treinamento = treinamento[colunas]
treinamento
```

```python
#alterando os nomes das colunas -> a data vai se chamar ds e o close se chamara y
treinamento.columns = ["ds", "y"]
treinamento
```

## Criando e treinando o modelo de Machine Learning 

```python
!pip install prophet --user
```

```python
from prophet import Prophet
```

```python
#criar modelo
modelo = Prophet()
```

```python
#treinar o modelo
modelo.fit(treinamento)
```

## Gerando Previsões

```python
modelo.make_future_dataframe(5)
```

```python
periodo = modelo.make_future_dataframe(90)
periodo.head()
```

```python
previsoes = modelo.predict(periodo)
previsoes.tail()
```

## Visualizando Dados

```python
from prophet.plot import plot_plotly
```

```python
graficos = plot_plotly(modelo, previsoes)
graficos.write_html("Previsões.html")
graficos.show()
```
